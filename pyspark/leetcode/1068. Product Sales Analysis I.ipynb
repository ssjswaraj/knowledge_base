```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.getOrCreate()
```

### Example 1

#### Input

**Sales table**

| sale_id | product_id | year | quantity | price |
|---------|------------|------|----------|-------|
| 1 | 100 | 2008 | 10 | 5000 |
| 2 | 100 | 2009 | 12 | 5000 |
| 7 | 200 | 2011 | 15 | 9000 |

**Product table**

| product_id | product_name |
|------------|--------------|
| 100 | Nokia |
| 200 | Apple |
| 300 | Samsung |

#### Output

| product_name | year | price |
|--------------|------|-------|
| Nokia | 2008 | 5000 |
| Nokia | 2009 | 5000 |
| Apple | 2011 | 9000 |

#### Explanation

- From `sale_id = 1`, Nokia was sold for `5000` in the year `2008`.
- From `sale_id = 2`, Nokia was sold for `5000` in the year `2009`.
- From `sale_id = 7`, Apple was sold for `9000` in the year `2011`.

```python
sales_data = [
    (1, 100, 2008, 10, 5000),
    (2, 100, 2009, 12, 5000),
    (7, 200, 2011, 15, 9000)
]

sales_df = spark.createDataFrame(
    sales_data,["sale_id", "product_id", "year", "quantity", "price"])

product_data = [
    (100, "Nokia"),
    (200, "Apple"),
    (300, "Samsung")
]

product_df = spark.createDataFrame(
    product_data,["product_id", "product_name"])

sales_df.show()

product_df.show()
```

```python
sales_df.alias("s").join(
    product_df.alias("p"),
     (col("s.product_id")==col("p.product_id")),
    "inner"
     ).select(col("p.product_name"),col("s.year"),col("s.price")).show()
```
